```python
import pandas as pd
from pathlib import Path
import json

# Carpeta donde guardaremos los archivos
output_dir = Path(".")
```

```python
ventas = pd.DataFrame({
    'producto': ['Laptop', 'Mouse', 'Teclado'],
    'cantidad': [10, 50, 30],
    'total': [9999.0, 1499.5, 1499.7]
})

productos = pd.DataFrame({
    'nombre': ['Laptop', 'Mouse', 'Teclado'],
    'precio': [999.9, 29.99, 49.99],
    'stock': [50, 200, 150]
})

clientes = pd.DataFrame({
    'nombre': ['Ana', 'Luis', 'Pedro'],
    'edad': [25, 30, 28],
    'ciudad': ['Santiago', 'Valparaíso', 'Concepción']
})

print("DataFrames creados ✅")
```

```python
excel_path = output_dir / "datos.xlsx"

with pd.ExcelWriter(excel_path) as writer:
    ventas.to_excel(writer, sheet_name="ventas", index=False)
    productos.to_excel(writer, sheet_name="productos", index=False)
    clientes.to_excel(writer, sheet_name="clientes", index=False)

print(f"Excel guardado: {excel_path}")
print(f"Tamaño: {excel_path.stat().st_size} bytes")
```

```python
for hoja in ["ventas", "productos", "clientes"]:
    df = pd.read_excel(excel_path, sheet_name=hoja)
    print(f"\nHoja: {hoja}")
    print(df)
```

```python
json_path = output_dir / "ventas.json"

ventas.to_json(json_path, orient='records', indent=2)

print(f"JSON guardado: {json_path}")
print(f"Tamaño: {json_path.stat().st_size} bytes")
print("\nContenido:")
print(json_path.read_text())
```

```python
df_desde_json = pd.read_json(json_path, orient='records')
print("DataFrame desde JSON:")
print(df_desde_json)
```

```python
csv_path = output_dir / "ventas.csv"
ventas.to_csv(csv_path, index=False)

print("Comparación de tamaños para los mismos datos:")
print(f"  CSV:   {csv_path.stat().st_size} bytes")
print(f"  JSON:  {json_path.stat().st_size} bytes")
print(f"  Excel: {excel_path.stat().st_size} bytes")
```

## Conclusiones

- **CSV** es el mejor formato para análisis — liviano y universal
- **JSON** es ideal para APIs y datos web
- **Excel** es para reportes humanos, no para análisis masivo
- Para datasets grandes (millones de filas) usar **Parquet** — más rápido que todos
